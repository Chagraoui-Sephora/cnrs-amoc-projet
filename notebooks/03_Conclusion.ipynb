### Conclusion 
Ce projet avait pour objectif d'évaluer l'influence de la résolution spatiale sur la représentation de plusieurs proxys de surface associés à l'AMOC. L'analyse montre que la diminution de résolution préserve les structures de grande échelle, mais atténue progressivement les fronts océaniques, les gradients les plus intenses et la variabilité de méso-échelle. Ces résultats soulignent l'importance d'une résolution élevée pour représenter fidèlement les structures fines de l'Atlantique Nord et quantifier la diminution de l'AMOC
 

### Limites identifiées lors de cette étude : 
 
- La simulation basse résolution a été obtenue par dégradation spatiale (lissage) de la simulation 1/12° et non à partir d’une simulation numérique intégrée directement à 1/4°. Cette approche isole l’effet du changement de résolution sur les diagnostics, mais ne reproduit pas les modifications de la dynamique océanique induites par une simulation réellement calculée à basse résolution. Les résultats doivent donc être interprétés comme une première évaluation de l’effet du lissage spatial sur la représentation des fronts océaniques.
 
- L’étude s’est limitée aux variables de surface (SST, SSS et SSH), utilisées comme proxys de la circulation océanique.
- Les diagnostics ont principalement porté sur la représentation des fronts océaniques et de leur variabilité spatiale.
- Les analyses restent descriptives et ne permettent pas d’identifier directement les mécanismes physiques responsables des différences observées entre les résolutions.

### Axes d'amélioration : 
 
- Étudier directement l’AMOC, son évolution  et la  structure tridimensionnelle de l’océan en fonction de la résolution plutôt que de se limiter à ses proxys de surface.
- Quantifier les transports de chaleur, de sel et de masse associés aux différentes résolutions.
- Analyser plus finement l’activité de méso-échelle (tourbillons, méandres, énergie cinétique turbulente) et son impact sur la représentation des fronts.
